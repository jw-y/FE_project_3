```python
today = date(2021, 6, 4)
expir_date = date(2021, 6, 11)
#tau = BS_functions.cal_tau(today, expir_date)
S = 4229.89 #2021-06-04 close
```

```python
imvol_list = []
for idx, row in df.iterrows():
    contractSym = row.contractSymbol
    lastTradeDate = row.lastTradeDate
    premium = row.lastPrice
    strike = row.strike
    
    print(premium, strike)
    
    if lastTradeDate.year!=2021 or lastTradeDate.month!=6 or lastTradeDate.day!=4:
        continue
    
    if contractSym[3]=='W':
        tau = BS_functions.cal_tau(today, expir_date)
        print('hi')
    else:
        tau = BS_functions.cal_tau(today, expir_date, PM_settle=False)
    
    x = BS_functions.call_imvol(premium, S, strike, tau, rf_rate)
    if x:
        m = BS_functions.cal_moneyness(S, strike, tau, x, rf_rate)
        imvol_list.append((m, x))
```
